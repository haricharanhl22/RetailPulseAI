```python
import pandas as pd

# Read with comma separator (correct), then clean up trailing semicolons
df = pd.read_csv("../data/superstore.csv", encoding="latin1", sep=",")

# Strip whitespace AND trailing semicolons from column names
df.columns = df.columns.str.strip().str.rstrip(";")

# Also clean the last column's values (they may have trailing ;)
last_col = df.columns[-1]
df[last_col] = df[last_col].astype(str).str.rstrip(";").astype(float)

print("✅ Fixed! Columns:", df.columns.tolist())
print("Shape:", df.shape)
df.head()
```

```python
# Fix: convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Now this will work
monthly_profit = df.groupby(
    df["Order Date"].dt.to_period("M")
)["Profit"].sum()

monthly_profit.plot(figsize=(12, 5), title="Monthly Profit")
```

```python
average_order_value = df["Sales"].mean()

profit_margin = (
    df["Profit"].sum() / df["Sales"].sum()
) * 100

print("Average Order Value:", round(average_order_value, 2))
print("Profit Margin:", round(profit_margin, 2), "%")
```

```python
from sklearn.ensemble import IsolationForest
```

```python
anomaly_data = df[["Sales", "Profit"]]
```

```python
model = IsolationForest(
    contamination=0.02,
    random_state=42
)

df["Anomaly"] = model.fit_predict(anomaly_data)
```

```python
anomalies = df[df["Anomaly"] == -1]

anomalies[["Sales", "Profit"]].head()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Sales"],
    df["Profit"],
    alpha=0.5
)

plt.scatter(
    anomalies["Sales"],
    anomalies["Profit"],
    alpha=0.8
)

plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Anomaly Detection")

plt.show()
```
